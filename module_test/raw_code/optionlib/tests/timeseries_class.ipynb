```python
from abc import ABC, abstractmethod
class TimeSeriesBase(ABC):
    """
    Base class for time series data.
    """

    @abstractmethod
    def get_data(self):
        """
        Returns the time series data.
        """
        pass

    @abstractmethod
    def get_start_time(self):
        """
        Returns the start time of the time series.
        """
        pass

    @abstractmethod
    def get_end_time(self):
        """
        Returns the end time of the time series.
        """
        pass

    @abstractmethod
    def get_frequency(self):
        """
        Returns the frequency of the time series.
        """
        pass
```
